```python
import requests as rq
import bs4
```

```python
# The URL to get the text data from.
res = rq.get("https://dantri.com.vn/an-sinh/ngay-doan-tu-sau-nua-the-ky-nho-chuoi-gen-vo-hinh-20240718143043606.htm")
```

```python
res.status_code
```

```python
my_header = dict(res.headers)
my_header.keys()
```

```python
import requests
print(my_header.get('date', 'Không có thông tin về ngày'))
print(my_header.get('content-type', 'Không có thông tin về loại nội dung'))
print(my_header.get('content-language', 'Không có thông tin về ngôn ngữ nội dung'))
print(my_header.get('content-length', 'Không có thông tin về độ dài nội dung'))
```

```python
print(res.text)
```

```python
soup = bs4.BeautifulSoup(res.text, 'html.parser')
print(soup.prettify())
```

```python
# Join the text content from all the paragraphs.
x=soup.find_all('p')
n = len(x)
paragraphs = ''
for i in range(n):
    paragraphs += x[i].text.strip() + '\n'
```

```python
print(paragraphs)
```
